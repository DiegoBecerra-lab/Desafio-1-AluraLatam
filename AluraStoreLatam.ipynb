# Informe Final: Recomendación de Venta de Tienda – Alura Store

## Introducción

El presente informe tiene como objetivo **sintetizar los resultados del análisis de datos realizado sobre las cuatro tiendas de la cadena Alura Store**, con el fin de apoyar al Sr. Juan en la decisión de **qué tienda vender para iniciar un nuevo emprendimiento**.

El análisis se basó en indicadores clave de desempeño, tales como **ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio de los clientes, productos con mayor y menor volumen de ventas y el costo promedio de envío**. Asimismo, se utilizaron distintas visualizaciones (gráficos de barras, gráficos circulares y comparaciones entre métricas) para facilitar la interpretación de los datos y respaldar la toma de decisiones.

El enfoque principal del estudio es identificar la **tienda con menor eficiencia relativa dentro del conjunto**, es decir, aquella cuyo aporte al desempeño global de la cadena es menor y cuya venta representaría el menor impacto negativo para el negocio.

---

## Desarrollo del análisis

### 1. Ingresos totales por tienda

La facturación total permite evaluar el peso económico de cada tienda dentro de la cadena:

* **Tienda 1:** $1,150,880,400

* **Tienda 2:** $1,116,343,500

* **Tienda 3:** $1,098,019,600

* **Tienda 4:** $1,038,375,700

A partir de la visualización comparativa de ingresos, se observa que la **Tienda 4 presenta la menor facturación total**, lo que indica un menor volumen de ventas y una contribución más baja al ingreso global de Alura Store. En contraste, la Tienda 1 lidera claramente en este indicador.

---

### 2. Calificaciones promedio de los clientes

Las calificaciones reflejan la percepción de los clientes sobre su experiencia de compra:

* **Tienda 1:** 3.98

* **Tienda 2:** 4.04

* **Tienda 3:** 4.05

* **Tienda 4:** 4.00

Las diferencias entre tiendas no son extremas, pero se aprecia que la **Tienda 3 cuenta con la mejor valoración promedio**, seguida de la Tienda 2. La Tienda 4 mantiene una calificación aceptable y cercana al promedio general, mientras que la Tienda 1 presenta la calificación más baja.

Este resultado indica que la Tienda 4 **no destaca negativamente en satisfacción del cliente**, aunque tampoco lidera este indicador.

---

### 3. Costo promedio de envío

El costo de envío es un factor relevante tanto para la experiencia del cliente como para la eficiencia operativa:

* **Tienda 1:** $26,018.61

* **Tienda 2:** $25,216.24

* **Tienda 3:** $24,805.68

* **Tienda 4:** $23,459.46

Según el gráfico comparativo, la **Tienda 4 presenta el costo de envío promedio más bajo**, lo que representa una fortaleza operativa. Sin embargo, este beneficio no se traduce en mayores ingresos, lo que sugiere que, a pesar de su eficiencia logística, su escala de operación es menor.

---

### 4. Productos más y menos vendidos

* **Tienda 1** concentra sus productos más vendidos en electrodomésticos y electrónicos (microondas, TV LED, secadoras), aunque presenta bajas ventas en algunos productos tecnológicos y educativos.

* **Tienda 2** muestra un portafolio más diverso, destacando libros, instrumentos musicales y electrónicos entre sus productos más vendidos.

* **Tienda 3** tiene un claro liderazgo en productos de la categoría muebles (camas, mesas y kits), lo que indica una especialización exitosa.

* **Tienda 4** presenta buenos niveles de venta en muebles y algunos productos específicos, pero también muestra debilidades en artículos clave como refrigeradores y ciertos productos tecnológicos, lo cual limita su impacto en la facturación total.

Los gráficos de productos más y menos vendidos evidencian que la **Tienda 4 no lidera en ningún producto de alto impacto**, reforzando su posición como la tienda con menor peso estratégico.

---

### 5. Categorías de productos más y menos vendidas

En las cuatro tiendas, la categoría **Muebles** es consistentemente la más popular, seguida por **Electrónicos** y **Juguetes**.

No obstante, al analizar la distribución por tienda, se observa que:

* La **Tienda 3** destaca por un mayor volumen en muebles.
* La **Tienda 1 y 2** mantienen un desempeño sólido en electrónicos y electrodomésticos.
* La **Tienda 4** presenta menores volúmenes en categorías clave como **Electrodomésticos** y **Artículos para el hogar**, categorías que suelen aportar ingresos relevantes.

Esto indica que, aunque la Tienda 4 tiene una distribución relativamente equilibrada, su **volumen total por categoría es inferior al de las demás tiendas**.

---

## Conclusión y recomendación final

Con base en el análisis integral de los datos y en las visualizaciones generadas, se concluye que la **Tienda 4 es la mejor opción para ser vendida**.

### Justificación de la decisión:

* Es la **tienda con menor facturación total**, lo que implica un menor aporte económico a la cadena.
* No lidera en calificaciones de clientes ni en volumen de ventas por producto o categoría.
* Aunque presenta el **menor costo de envío promedio**, esta ventaja operativa no compensa su bajo impacto en ingresos.
* Su venta representaría el **menor riesgo estratégico**, ya que las Tiendas 1, 2 y 3 muestran un desempeño más sólido en términos de ingresos, especialización por categorías y volumen de ventas.

En conclusión, vender la **Tienda 4 permitiría al Sr. Juan liberar recursos para su nuevo emprendimiento sin afectar de manera significativa el rendimiento global de Alura Store**, manteniendo dentro de la cadena aquellas tiendas que generan mayor valor y presentan mejores perspectivas de crecimiento.
